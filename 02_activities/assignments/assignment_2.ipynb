#### Starter Code for Question 1

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right
# BFS inputs
# [1, 2, 2, 3, 5, 6, 7] 2
# [1, 10, 2, 3, 10, 12, 12] 10
# [10, 9, 8, 7] -1
def is_duplicate(root: TreeNode) -> int:
  return bfs_traversal(root, set())

def bfs_traversal(node: TreeNode, seen: set) -> int:
  if not node:
    return -1
  
  if node.val in seen:
    return node.val
  
  seen.add(node.val)

  left_result = bfs_traversal(node.left, seen)
  if left_result != -1:
    return left_result
  
  return bfs_traversal(node.right, seen)
```

### Your answer here
My Partner has been assigned to Question 2, which is to determine if the input string is a valid bracket sequence or not. 
The input string will contain only the characters '(', ')', '{', '}', '[' and ']'. 
The output should be a boolean value indicating whether the input string is a valid bracket sequence or not. 
Open brackets should be closed in the correct order, and all brackets should be closed.


-   Create 1 new example that demonstrates you understand the problem. Trace/walkthrough 1 example that your partner made and explain it.


```python
# Your answer here
str = "([]({}))" # returns: True

# Partner's example:
Input: "{[()()]}"  
Output: True  
```

### Walkthrough based on the partner's example:
Input: "{[()()]}"
Output: True

Step 1: stack = []
Read '{' → open bracket, push to stack → stack = ['{']
Read '[' → open bracket, push to stack → stack = ['{', '[']
Read '(' → open bracket, push to stack → stack = ['{', '[', '(']
Read ')' → close bracket, matches '(', pop → stack = ['{', '[']
Read '(' → open bracket, push to stack → stack = ['{', '[', '(']
Read ')' → close bracket, matches '(', pop → stack = ['{', '[']
Read ']' → close bracket, matches '[', pop → stack = ['{']
Read '}' → close bracket, matches '{', pop → stack = []

At the end, the stack is empty, so all brackets are matched and closed in the correct order.
Therefore, the function returns True.


-   Copy the solution your partner wrote. 


```python
# Your answer here

def is_valid_brackets(s: str) -> bool:
    stack = []
    bracket_map = {')': '(', ']': '[', '}': '{'}
    for char in s:
        if char in bracket_map.values():
            stack.append(char)
        elif char in bracket_map:
            if not stack or stack.pop() != bracket_map[char]:
                return False
    return not stack
```

```python
# Assertion and more test cases
assert is_valid_brackets(str) == True, "Test case failed"
assert is_valid_brackets("([)]") == False, "Test case failed"
assert is_valid_brackets("{[()]}") == True, "Test case failed"
```


-   Explain why their solution works in your own words.


### Your answer here
The solution works because it uses a stack to keep track of opening brackets as it scans the string from left to right.
When it encounters an opening bracket ('(', '[', or '{'), it pushes it onto the stack.
When it encounters a closing bracket (')', ']', or '}'), it checks if the stack is not empty and if the top of the stack matches the corresponding opening bracket.
If it matches, it pops the opening bracket from the stack; if not, it returns False immediately.
At the end, if the stack is empty, it means all brackets were matched and closed in the correct order, so it returns True.
This approach ensures that every opening bracket has a matching closing bracket and that the pairs are properly nested.


-   Explain the problem’s time and space complexity in your own words.


### Your answer here
The time complexity of the solution is O(n), where n is the length of the input string.
This is because each character in the string is processed exactly once.
The space complexity is also O(n) in the worst case, which happens when all characters are opening brackets and are pushed onto the stack.
In summary, both time and space complexity are linear with respect to the length of the input string.
